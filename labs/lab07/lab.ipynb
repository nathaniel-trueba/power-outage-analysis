<br>

#### `hashtag_list`

First, complete the implementation of the function `hashtag_list`, which takes in a Series of tweet texts and returns a Series containing a list of hashtags present in each tweet's text. If a tweet's text doesn't contain a hashtag, the Series should contain an empty list for that tweet. Don't include the leading `'#'` symbol as part of the hashtag.

A hashtag is any string without whitespace that immediately follows a `'#'` symbol. To clarify, a hashtags can contain a `'#'` symbol, as long as there is no whitespace. For example, `'#data#science'` is counted as a single hashtag, whereas `'#data #science'` is considered two separate hashtags. Remember not to include the leading `'#'` symbol in the hashtag.

<br>

#### `most_common_hashtag`

Complete the implementation of the function `most_common_hashtag`, which takes in a Series of hashtag lists (as is outputted by `hashtag_list`) and returns a Series consisting of a single hashtag per tweet, as follows: 
- If the tweet has no hashtags, the entry should in the output Series should be `NaN`.
- Otherwise, the entry should be the hashtag from that tweet that appears the most frequently **in the whole input Series**. If there is a tie for the most common, any of the most common can be returned.

For example, if the input Series was `pd.Series([["news"], ["news"], ["sports", "news", "sports"]])`, the output would be `pd.Series(["news", "news", "news"])`. Even though `"sports"` was more common in the third list than `"news"`, `"news"` is more common than `"sports"` among all hashtags in the Series overall.

***Hint:*** The Series method [`.explode()`](https://pandas.pydata.org/docs/reference/api/pandas.Series.explode.html) is convenient here, but you can also do this problem without it by simply summing lists to combine them into one longer list.

<br>

#### `create_features`

Finally, complete the implementation of the function `create_features`, which takes in a DataFrame `tweets` that has just a single column, `'text'`, and returns a DataFrame with the same index as `tweets` (i.e. the rows correspond to the same tweets) and the following columns:
* `'num_hashtags'`, the number of hashtags present in the tweet.
* `'common_hashtag'`, the most common hashtag used within the tweet (using the result of `most_common_hashtag` above).
* `'num_tags'`, the number of tags present in the tweet.

    - A hyperlink is a string starting with `'http://'` or `'https://'`, not followed by whitespaces.

    - A retweet is a tweet that begins with `'RT'`.
* `'text'`, a version of the tweet's text that is cleaned according to the following steps, **in this exact order**:
    1. All hashtags, tags, hyperlinks, and instances of `'RT'` should be replaced with a single space.
